# How do we verify the users credentials are correct?  

1) User is gonna hit the login endpoint of API and gonna provide the username (email) and password (plain text).
2) And API is gonna find the user from the db.
3) The database is gonna send back the information about the user including the password (hashed) not plain password.
4) Once the plain text password is hashed we cant get the original password (hashing is one way). If you want to retrieve the password back we use encryption which is two way
5) To verify the user password we take the hashed password stored in the db and compare it with the password that is user given after converting the user given password into hash using hashing algorithm
6) If equal GET IN!!!!!!!!
